```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_datasets = load_breast_cancer()
df = pd.DataFrame(cancer_datasets['data'], columns=cancer_datasets['feature_names'])
df.head()
```

```python
## doing standardization to get all the data on same scale 
## stnadardization reduce the dominance of bigger value in the dataset

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
print(scaled_data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
pca_scaled_data = pca.transform(scaled_data)
```

```python
pca_scaled_data
```

```python
pca.explained_variance_
## pc1 has magnitude of 13.4 
## pc2 has magnitude of 5.70
```

```python
plt.figure(figsize=(10,10))
plt.scatter(pca_scaled_data[:,0], pca_scaled_data[:,1], c=cancer_datasets['target'], cmap='plasma')
plt.xlabel('PC1')
plt.ylabel('PC2')
```
